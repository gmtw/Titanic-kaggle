# Matrices de confusión y Curvas ROC

```python
X_train, X_test, Y_train, Y_test =  train_test_split(X,Y,test_size = 0.3, random_state = 0)
```

```python
lm = linear_model.LogisticRegression()
lm.fit(X_train,Y_train)
```

```python
probs = lm.predict_proba(X_test)
```

```python
prob = probs[:,1]
prob_df = pd.DataFrame(prob)
threshold = 0.03
prob_df['prediction']=np.where(prob_df[0]>=threshold,1,0)
prob_df['actual'] = list(Y_test)
prob_df
```

```python
confusion_matrix = pd.crosstab( prob_df.prediction, prob_df.actual)
confusion_matrix
```

```python
TN = confusion_matrix[0][0]
TP = confusion_matrix[1][1]
FN = confusion_matrix[0][1]
FP = confusion_matrix[1][0]
```

```python
sens = TP/(TP+FN)
sens
```

```python
espc_1 = 1-TN/(TN+FP)
espc_1
```

```python
threshold =[0.03,0.04, 0.05, 0.07, 0.10, 0.12, 0.15, 0.18, 0.25, 0.3,0.4,0.5]
sensitivities = [1]
especifities_1 = [1]
for t in threshold:
    prob_df['prediction']=np.where(prob_df[0]>=t,1,0)
    prob_df['actual'] = list(Y_test)
    prob_df.head(10)
    
    confusion_matrix = pd.crosstab( prob_df.prediction, prob_df.actual)
    TN = confusion_matrix[0][0]
    TP = confusion_matrix[1][1]
    FP = confusion_matrix[0][1]
    FN = confusion_matrix[1][0]
    
    sens = TP/(TP+FN)
    sensitivities.append(sens)
    espc_1 = 1-TN/(TN+FP)
    especifities_1.append(espc_1)
    
sensitivities.append(0)
especifities_1.append(0)
```

```python
sensitivities
```

```python
especifities_1
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.plot(especifities_1, sensitivities, marker ='o', linestyle = '--', color = 'red')
x = [i*0.01 for i in range(0,100)]
y = [i*0.01 for i in range(0,100)]
plt.plot(x,y)
plt.xlabel('1-Especifidad')
plt.ylabel('Sensitividad')
plt.title('curva ROC')
plt.show()
```
